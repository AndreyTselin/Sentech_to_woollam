```python
import numpy as np
import pandas as pd


def Sentech2Woollam(path):
    df = pd.read_table(path, index_col=False)

    col = df.columns
    angle_number = len(col)//2
    col_len = len(df[col[0]])


    # Create first rows of 1st column

    head_1 = np.array(['VASEmethod[EllipsometerType=0, Isotropic, AutoRetarder=0, TrackPol=1, Pol>=0, Revs=10.0, AutoSlit=1700, WVASE=3.916, HardVer=6.256, Fri Mar 25 15:09:27 2022]',
                     'Original[2022_03_25_ws2_graphene_sio2_300nm_si_50-70deg_by5deg_290-3300nm.dat]',
                     'nm'])

    # Create first rows of all column except 1st

    head_2 = np.array([' ', ' ', ' '])



    # Create a 1st column that contains wavelengthes

    Wl = pd.Series(head_1)
    for i in range(angle_number):
        Wl = pd.concat((Wl, df[col[0]]), ignore_index=True)

    # Create a 2nd column that contains Angles

    Angle = pd.Series(head_2)
    for i in range(1, 2*angle_number, 2):
        ang = pd.Series(float(col[i]) * np.ones(col_len))
        Angle = pd.concat((Angle, ang), ignore_index=True)

    # Create a 3rd column that contains Psi

    Psi = pd.Series(head_2)
    for i in range(angle_number):
        Psi = pd.concat((Psi, df[col[i+1]]), ignore_index=True)

    # Create a 4th column that contains Delta

    Delta = pd.Series(head_2)
    for i in range(angle_number):
        Delta = pd.concat((Delta, df[col[i+2]]), ignore_index=True)

    # Create columns that contain errors

    err = 0.05 * np.ones(angle_number * col_len)
    Error = pd.concat((pd.Series(head_2), pd.Series(err)), ignore_index=True)

    # Create new df where data in required form

    df_1 = pd.DataFrame()
    df_1['1'] = Wl
    df_1['2'] = Angle
    df_1['3'] = Psi
    df_1['4'] = Delta
    df_1['5'] = Error
    df_1['6'] = Error

    # Create new path

    parts = path.split('.c')
    new_path = parts[0] + '_woollam_form.' + 'c' + parts[1]

    # Write new DataFrame

    df_1.to_csv(new_path, sep='\t', header=False, index=False)
    message = "if error try add r before path"
    print(message)
    return df_1
```
