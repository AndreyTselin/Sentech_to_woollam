```python
import numpy as np
import pandas as pd

def transform(path):
    df_1 = pd.read_table(path, index_col=False) 
    wl = np.array(df_1['WAVELENGTH'])
    #psi_40 = np.array(df_1['39.96000'])
    psi_50 = np.array(df_1['49.96000'])
    psi_60 = np.array(df_1['59.96000'])
    psi_70 = np.array(df_1['69.96000'])
    #delta_40 = np.array(df_1['39.96000.1'])
    delta_50 = np.array(df_1['49.96000.1'])
    delta_60 = np.array(df_1['59.96000.1'])
    delta_70 = np.array(df_1['69.96000.1'])
    head_1 = np.array(['VASEmethod[EllipsometerType=0, Isotropic, AutoRetarder=0, TrackPol=1, Pol>=0, Revs=10.0, AutoSlit=1700, WVASE=3.916, HardVer=6.256, Fri Mar 25 15:09:27 2022]',
                 'Original[2022_03_25_ws2_graphene_sio2_300nm_si_50-70deg_by5deg_290-3300nm.dat]',
                 'nm'])
    head_2 = np.array([' ', ' ', ' '])
    Wl = np.concatenate((head_1, wl, wl, wl))
    Psi = np.concatenate((head_2, psi_50, psi_60, psi_70))
    Delta = np.concatenate((head_2, delta_50, delta_60, delta_70))
    l = len(wl)
    L = len(Wl)
    #a_40  = 39.96000 * np.ones(l)
    a_50  = 49.96000 * np.ones(l)
    a_60  = 59.96000 * np.ones(l)
    a_70  = 69.96000 * np.ones(l)
    Angles = np.concatenate((head_2, a_50, a_60, a_70))
    Error = 0.05 * np.ones(L-3)
    Error_n = np.concatenate((head_2, Error))
    dat = np.array([Wl, Angles, Psi, Delta, Error_n, Error_n])
    dat_1 = np.transpose(dat)
    df_3 = pd.DataFrame(dat_1)
    parts = path.split('.')
    new_path = parts[0] + '_woollam_form.' + parts[1]
    df_3.to_csv(new_path, sep='\t', header=False, index=False)
    return df_3
```
